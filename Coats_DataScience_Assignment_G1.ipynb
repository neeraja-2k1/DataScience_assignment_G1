* we cannot use label encoder which is used for ordinal variables in this case.
* so we move on to one-hot or dummy encoding

* 1)label encoding:
    * When the categorical feature has some order or is ordinal. 
    * For example, an army’s position is ordinal and the highest position has a high number.
    * When the number of categories is large.
* 2)One-hot encoding:
    * When the categorical feature has no order or is not ordinal. 
    * For example, the city where a person is in is not ordinal 
    * so one-hot encoding can be used for such types of columns.When the number of unique categorical features is less. 
    * This is because more features increase the model’s complexity and training time.
* 3)Dummy encoding:
    * Dummy coding scheme is similar to one-hot encoding.
    * In the case of one-hot encoding, for N categories in a variable, it uses N binary variables. 
    * The dummy encoding is a small improvement over one-hot-encoding. 
    * Dummy encoding uses N-1 features to represent N labels/categories
    * The drop_first=True drops one column from the resulted dummy features. The purpose is to avoid multicollinearity.

## 6.1 ML technique 1 + Justification

* Linear models:

SMOTE is performed to handle imbalance target class by oversampling.
SMOTE (Synthetic Minority Over-sampling Technique) is a popular method used to address class imbalance in datasets, 
especially in binary classification problems where one class is significantly underrepresented compared to the other.
SMOTE helps in balancing the dataset by generating synthetic samples for the minority class.

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=42)
X_sm, y_sm = sm.fit_resample(X, y)
print(f'''Shape of X before SMOTE: {X.shape}
Shape of X after SMOTE: {X_sm.shape}''')
print('\nBalance of positive and negative classes (%):')
y_sm.value_counts(normalize=True) * 100
```

```python
X_sm
```

```python
y_sm
```

```python
##---------Type the code below this line------------------##
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm,test_size=0.20,shuffle=True,random_state=None)
#'Age','Type_Travel Agency', ,'Agency_Code_JZI','Product Name_Customised Plan'
#'Age','Commision','Duration','Sales','Type_Travel Agency','Agency_Code_JZI','Agency_Code_EPX','Product Name_Customised Plan','Product Name_Cancellation Plan', 'Product Name_Silver Plan'
```

```python
final_selected_features = ['Age','Duration','Sales','Type_Travel Agency','Agency_Code_EPX','Agency_Code_JZI','Product Name_Customised Plan','Product Name_Cancellation Plan', 'Product Name_Silver Plan']
final_X_train = X_train[final_selected_features]
final_X_test = X_test[final_selected_features]
```

```python
scaler = MinMaxScaler()
features = final_X_train.columns
final_X_train = scaler.fit_transform(final_X_train)
final_X_train = pd.DataFrame(final_X_train,columns=features)
final_X_test = scaler.transform(final_X_test)
final_X_test = pd.DataFrame(final_X_test,columns=features)
final_X_train.head()
```

```python
model_names = []
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
roc_auc_scores = []
```

```python
def train_and_evaluate_model(model):
    model.fit(final_X_train,y_train)
    y_pred = model.predict(final_X_test)
    print(classification_report(y_test, y_pred))
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    roc_auc = roc_auc_score(y_test,y_pred)
    print("Accuracy Score:",accuracy)
    print("Precision Score:",precision)
    print("Recall Score:",recall)
    print("F1 Score:",f1)
    print("ROC AUC Score:",roc_auc)
    ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
    plt.show()
    PrecisionRecallDisplay.from_predictions(y_test,y_pred)
    plt.show()
    model_names.append(str(model))
    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)
    roc_auc_scores.append(roc_auc)
```

```python
train_and_evaluate_model(LogisticRegression())
```

```python
train_and_evaluate_model(SVC(kernel='linear',gamma=2))
```
